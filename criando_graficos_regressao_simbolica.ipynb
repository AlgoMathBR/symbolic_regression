```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#==========================================
# 1 gráfico
x = np.linspace(-100, 100, 400)
y1 = x**2
y2 = 1.50119395960383 * x**2
std_dev = np.std(y1 - y2)

plt.figure(figsize=(4.5, 2.25))  # comprimento, altura

# Subplot 1
plt.subplot(1, 2, 1) #LINHA, COLUNA, POSICAO
sns.lineplot(x=x, y=y1, label='x²')
sns.lineplot(x=x, y=y2, label='SR*', linestyle='--')
plt.title('Bubblesort')
plt.xlabel('x')
plt.ylabel('y')

#==========================================

# 2 gráfico
x1 = np.linspace(0.0001, 100, 400)
y11 = x1 * np.log(x1)
y22 = 3.9986968 * x1 - 1.46676437242608

# Subplot 2
plt.subplot(1, 2, 2)
sns.lineplot(x=x1, y=y11, label='x log x')
sns.lineplot(x=x1, y=y22, label='SR*', linestyle='--')
plt.title('Heapsort')
plt.xlabel('x')
plt.ylabel('y')

#==========================================

plt.tight_layout()  # Ajusta o layout para evitar sobreposição

plt.savefig('graficos_lado_a_lado2.png')  # Salva a imagem
plt.show()
```
