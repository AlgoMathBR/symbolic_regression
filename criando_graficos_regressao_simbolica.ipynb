# Criando os gráficos

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
```

```python
x = np.linspace(-100, 100, 400)

y1 = x**2
y2 = 1.50119395960383*x**2

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='x²')
sns.lineplot(x=x, y=y2, label='1.50119395960383*x²', linestyle='--')

plt.title('Bubblesort')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```

```python
x = np.linspace(0.0001, 100, 400)

y1 = x * np.log(x)
y2 = 3.9986968*x - 1.46676437242608

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='x log x')
sns.lineplot(x=x, y=y2, label='3.9986968*x - 1.46676437242608', linestyle='--')

plt.title('Heapsort')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```

```python
x = np.linspace(0.0001, 100, 400)

y1 = np.log(x)
y2 = np.log(x)**2 + 13.992754

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='log x')
sns.lineplot(x=x, y=y2, label='log(x)**2 + 13.992754', linestyle='--')

plt.title('Binary Search')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```

```python
x = np.linspace(-100, 100, 400)

y1 = x
y2 = 2*x + 1.9999998

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='x')
sns.lineplot(x=x, y=y2, label='2*x + 1.9999998', linestyle='--')

plt.title('Linear Search')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```

```python
x = np.linspace(0.0001, 100, 400)

y1 = x**1.585
y2 = 2.6843011025431*np.array([math.ceil(val)**2 for val in (1.19255337641243*x - 0.5)])

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='x**1.585')
sns.lineplot(x=x, y=y2, label='2.6843011025431*ceiling(1.19255337641243*x - 0.5)**2', linestyle='--')

plt.title('Karatsuba')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```

```python
x = np.linspace(0.0001, 100, 400)

y1 = x**2.81
y2 = np.array([math.ceil(val**2*(val - 0.34646276)*math.ceil(2.8932767*val - 8.245815)**2 + 91.353319552452) for val in x])

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='x**2.81')
sns.lineplot(x=x, y=y2, label='ceiling(x0**2*(x0 - 0.34646276)*ceiling(2.8932767*x0 - 8.245815)**2 + 91.353319552452)', linestyle='--')

plt.title('Multiply Matrix Strassen')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```

```python
x = np.linspace(1, 30, 400)

y1 = 2**x
y2 = np.sqrt(x) + 5.57341960392382*np.sqrt(1/x)*np.exp(np.abs(x)/2) - 120.4865

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='2**x')
sns.lineplot(x=x, y=y2, label='sqrt(x0) + 5.57341960392382*sqrt(1/x0)*exp(Abs(x0)/2) - 120.4865', linestyle='--')

plt.title('Fibonacci')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```

```python
x = np.linspace(1, 30, 400)

y1 = 2**x
y2 = 0.0036318859*x**6 + 11.967872*x

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='2**x')
sns.lineplot(x=x, y=y2, label='0.0036318859*x**6 + 11.967872*x', linestyle='--')

plt.title('Tower Hanoi')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```

```python
x = np.linspace(-100, 100, 400)

y1 = 343174 + x
y2 = 247581.64*x - 12091663.0

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='m + x')
sns.lineplot(x=x, y=y2, label='247581.64*x - 12091663.0', linestyle='--')

plt.title('Boyer Moore')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```

```python
x = np.linspace(-100, 100, 400)

y1 = x + x*2
y2 = 13.9936639841862*(-x - 0.156347049268578)**2

plt.figure(figsize=(2, 2))

sns.lineplot(x=x, y=y1, label='m * x')
sns.lineplot(x=x, y=y2, label='13.9936639841862*(-x - 0.156347049268578)**2', linestyle='--')

plt.title('Levenshtein Distance')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Show both plots
plt.show()
```
